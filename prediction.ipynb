# combining fertilizer_prediction 1 and 2

```python
df = pd.read_csv('../datasets/fertilizer_prediction2.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(['Crop'], axis=1)
df
```

```python
label_encoder = LabelEncoder() 
df['Fertilizer']= label_encoder.fit_transform(df['Fertilizer']) 
df
```

```python
df_scaled = df
for column in df.columns:
    if column == 'pH':
        df_scaled[column] = df[column]
        continue
    df_scaled[column] = (df[column] - df[column].mean()) / df[column].std()	
df_scaled
```

```python
X = df_scaled[['Temperature','Humidity','Moisture','Fertilizer', 'N', 'P', 'K']]
y = df_scaled['pH']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
x = regr.predict(X)
mean_squared_error(x,y) 
```

# prediction of fertilizer for fertilizer_prediction3

```python
df = pd.read_csv("../datasets/fertilizer_prediction1.csv")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(['Crop', 'Temperature', 'Humidity', 'Moisture', 'pH'], axis=1)
df
```

```python
label_encoder = LabelEncoder() 
df['Fertilizer']= label_encoder.fit_transform(df['Fertilizer']) 
df
```

```python
df_z_scaled = df 
for column in df.columns[:-1]:
    df_z_scaled[column] = (df_z_scaled[column] - df_z_scaled[column].mean()) / df_z_scaled[column].std()	
df_z_scaled
```

```python
X = df_z_scaled[df_z_scaled.columns[:-1]]
y = df_z_scaled[df_z_scaled.columns[-1]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
accuracy = accuracy_score(y_pred, y_test)
print(f"Accuracy: {accuracy}")
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(X, y)
y_pred = random_forest.predict(X)
accuracy = accuracy_score(y_pred, y)
print(f"Accuracy: {accuracy}")
```
